#### Replacing missing values

```python
list_values =['LotFrontage','GarageType','GarageYrBlt','GarageCond', 'GarageQual', 'GarageFinish', 'BsmtFinType2','BsmtExposure', 'BsmtQual','BsmtCond','BsmtFinType1','MasVnrType','MasVnrArea','Electrical']

def max_subcategory(column_names):
    global sub_categories
    sub_categories = []
    for name in column_names:
        for index, value in train.groupby([name])['Id'].nunique().items():
            if value == train.groupby([name])['Id'].nunique().max():
                sub_categories.append(index)
    return sub_categories

max_subcategory(list_values)
```

```python
def replace_missingvalues(list_values):
    for name in list_values:
        for sub in sub_categories:
            train[name].fillna(sub, inplace = True)
    return train.isnull().sum().sort_values(ascending= False).head(20)
replace_missingvalues(list_values)
```

Missing values are replaced with the value with more frecuency in the subcategories

## Hypothesis 2

### Determine if SalePrice is associated with at least 4 categorical variables in the dataset

### Evaluating Correlation between Categorical variables

Calculating Statistical Correlation:

Conducting Chi square test to establish whether the SalePrice and categorical  variables are correlated.
* Dependent Variable: SalePrice_range
* Independent Variable: Categorical Variables

Ho: The variables SalePrice_range and categorical variables are independent from each other among all subjects in the population.

Ha: The variables SalePrice_range and categorical variables are not independent from each other among all subjects in the population.

```python
def frecuencytable(cat_train, valid_corr = True):
    global column_list
    column_list = []
    for column in cat_train:
        freq_table = pd.crosstab(train['SalePrice_range'], cat_train[column])
        chi2_stat, p_val, dof, ex = stats.chi2_contingency(freq_table)
        if p_val <= 0.05 if valid_corr else p_val > 0.05:
            column_list.append(column)
            print(column)
            print("Chi square value is ",chi2_stat)
            print("P value is",p_val)
            print("Degrees of Freedom:",dof, "\n")
    return column_list

# Frecuency table of Sex and Survived
frecuencytable(cat_train, True)
```

```python
column_list # The variables that are dependent with SalePrice_range
```

The top 4 variables with more correlation are: Neighborhood, ExterQual, KitchenQual and BsmtQual.
